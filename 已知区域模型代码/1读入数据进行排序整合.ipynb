```python
import pandas as pd
import numpy as np

import os
os.environ["KMP_DUPLICATE_LIB_OK"]="TRUE"
```

```python
# 指定Excel文件路径
file_path = 'OriginData/涌水量数据20241015.xlsx'

# 读取Excel文件中的所有sheets
datas = pd.read_excel(file_path, sheet_name=None)
```

```python
datas
```

```python
# 将所有sheets合并成一个DataFrame
df = pd.concat(datas.values())
```

```python
df
```

```python
 # 要替换的经纬度点
 target_longitude = [103.99983208,103.99902164,103.99283195,103.99758843,103.99784393,103.99942942]
 target_latitude = [27.51751954,27.51609454,27.51540273,27.51294739,27.51373011,27.51555221]

 # 新的大地坐标
 new_easting = [401223.55202111,401142.234106562,400530.274579171,400093.2536,400119.1581,400277.2726]
 new_northing = [3044155.75568918,3043998.54415792,3043926.85739021,3044617.46577,3044703.6545,3044902.0639]

 # 找到匹配的行并替换经纬度
 for i in range(len(target_latitude)):
     df.loc[(df['横坐标'] == target_longitude[i]) & (df['纵坐标'] == target_latitude[i]), ['横坐标', '纵坐标']] = [new_easting[i], new_northing[i]]
```

```python
# 确保时间列是datetime类型
df['日期'] = pd.to_datetime(df['日期'], format='%Y/%m/%d')

# 按横坐标、纵坐标、高程进行排序，然后按时间标签升序排序
df_sorted = df.sort_values(by=['横坐标', '纵坐标', '高程', '日期'])

# 将日期时间列转换为仅包含日期的字符串
df_sorted['日期'] = df_sorted['日期'].dt.strftime('%Y/%m/%d')

# 保存排序后的DataFrame到新的Excel文件
output_file_path = 'OutputData/sorted_data.xlsx'
df_sorted = df_sorted.dropna()
df_sorted.to_excel(output_file_path, index=False)
```
