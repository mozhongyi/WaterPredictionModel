```python
import pandas as pd

# 读取Excel文件（假设列名在第一行）
data = pd.read_excel('OriginData/df_水位水质层面断层.xlsx',index_col = 0)
```

```python
df = data
df
```

```python
import re
df.columns = df.columns.str.strip().str.replace(r"[\s_\-]+", "", regex=True)

# 使用正则表达式匹配列名
pattern = re.compile(r"^层面(\.\d+)?$")
columns_to_drop_ceng = [col for col in df.columns if pattern.match(str(col))]

# 删除匹配的列
df = df.drop(columns=columns_to_drop_ceng)

# 使用正则表达式匹配列名
pattern = re.compile(r"^断层(\.\d+)?$")
columns_to_drop_duan = [col for col in df.columns if pattern.match(str(col))]

# 删除匹配的列
df = df.drop(columns=columns_to_drop_duan)

# 刪除'日期'的列
df = df.drop(columns = ['日期'])

# 保存结果
df.to_excel('OutputData/df_水位水质层面断层_full.xlsx', index=False)
```
